```python
# Ej 1

import pandas as pd

url = 'https://raw.githubusercontent.com/fpineyro/homework-0/master/heart_disease.csv'
d_hd = pd.read_csv(url)

display(d_hd.head())
```

```python

# Ej 2

display(d_hd.head(5))
display(d_hd.tail(5))
```

```python
# Ej 3

seleccion = d_hd[['age', 'has_heart_disease']]
display(seleccion)
```

```python
# Ej 4

filtro = d_hd[d_hd['serum_cholestoral'] > 250]
display(filtro)
```

```python
# Ej 5

nulos = d_hd.isnull().sum()
display(nulos)
```

```python
# Ej 6

d_hd['thal'] = d_hd['thal'].fillna(999)

display(d_hd[d_hd['thal'] == 999])
```

```python
# Ej 7

filtro_edad_colesterol = d_hd[(d_hd['age'] > 60) & (d_hd['serum_cholestoral'] > 250)]
display(filtro_edad_colesterol.head(10))
```

```python
# Ej 8

slope_modificada = d_hd['slope'].replace({1: 'ascendente', 2: 'descendente'})
display(slope_modificada)
```

```python
# Ej 9

d_hd['is_healthy'] = (d_hd['heart_disease_severity'] == 0)
display(d_hd.head(22))
```

```python
# Ej 10

d_hd_clean = d_hd.copy()
d_hd_clean = d_hd_clean.dropna()

display(d_hd_clean.isnull().sum())
```
